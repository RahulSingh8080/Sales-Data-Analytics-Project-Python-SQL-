```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Create a Ticker object for Tesla
tesla = yf.Ticker("TSLA")

# Extract historical stock information
tesla_data = tesla.history(period="max")

# Reset the index
tesla_data.reset_index(inplace=True)

# Display the first five rows of the dataframe
print(tesla_data.head())

# URL of the webpage to download
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Save the text of the response as a variable named 'html_data'
html_data = response.text

# Parse the html_data with BeautifulSoup
soup = BeautifulSoup(html_data, 'html.parser')

# Find the table with 'Tesla Revenue'
table = soup.find('table', {'id': 'tesla-revenue'}) # Update the table id if needed

# Extract headers
headers = [header.text for header in table.find_all('th')]

# Extract rows
rows = []
for row in table.find_all('tr'):
    rows.append([cell.text for cell in row.find_all('td')])

# Create the dataframe
tesla_revenue = pd.DataFrame(rows, columns=headers)

# Rename columns to 'Date' and 'Revenue'
tesla_revenue.columns = ['Date', 'Revenue']

# Display the first few rows of the dataframe
print(tesla_revenue.head())
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# URL of the webpage to download
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Save the text of the response as a variable named 'html_data'
html_data = response.text

# Parse the html_data with BeautifulSoup
soup = BeautifulSoup(html_data, 'html.parser')

# Initialize the tesla_revenue variable
tesla_revenue = None

# Find all tables in the HTML
tables = soup.find_all('table')

# Extract the Tesla Revenue table
for table in tables:
    headers = [header.text for header in table.find_all('th')]
    if 'Date' in headers and 'Revenue' in headers:
        rows = []
        for row in table.find_all('tr'):
            cells = [cell.text.strip() for cell in row.find_all('td')]
            if cells:
                rows.append(cells)
        tesla_revenue = pd.DataFrame(rows, columns=['Date', 'Revenue'])
        break

# Check if the tesla_revenue variable has been populated
if tesla_revenue is not None:
    # Display the first few rows of the dataframe
    print(tesla_revenue.head())
else:
    print("Tesla Revenue table not found.")
```

Question 5 - Tesla Stock and Revenue Dashboard - 2 Points

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensuring matplotlib plots inline (if you're using Jupyter Notebook)
%matplotlib inline
```

```python
import yfinance as yf
import requests
from bs4 import BeautifulSoup

# Create a Ticker object for Tesla
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)

# URL of the webpage to download
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"

# Send an HTTP GET request to the URL
response = requests.get(url)
html_data = response.text

# Parse the html_data with BeautifulSoup
soup = BeautifulSoup(html_data, 'html.parser')
tables = soup.find_all('table')

# Initialize the tesla_revenue variable
tesla_revenue = None
for table in tables:
    headers = [header.text for header in table.find_all('th')]
    if 'Date' in headers and 'Revenue' in headers:
        rows = []
        for row in table.find_all('tr'):
            cells = [cell.text.strip() for cell in row.find_all('td')]
            if cells:
                rows.append(cells)
        tesla_revenue = pd.DataFrame(rows, columns=['Date', 'Revenue'])
        break

# Convert 'Date' to datetime and 'Revenue' to numeric
tesla_revenue['Date'] = pd.to_datetime(tesla_revenue['Date'])
tesla_revenue['Revenue'] = tesla_revenue['Revenue'].replace({'\$': '', ',': ''}, regex=True).astype(float)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(tesla_data['Date'], tesla_data['Close'], label='Closing Price')
plt.title('Tesla Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Stock Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=tesla_revenue, x='Date', y='Revenue', marker='o')
plt.title('Tesla Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue (USD in Millions)')
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(12, 10))

# Plotting Tesla Stock Price
ax[0].plot(tesla_data['Date'], tesla_data['Close'], label='Closing Price', color='blue')
ax[0].set_title('Tesla Stock Price Over Time')
ax[0].set_xlabel('Date')
ax[0].set_ylabel('Stock Price (USD)')
ax[0].legend()
ax[0].grid(True)

# Plotting Tesla Revenue
sns.lineplot(data=tesla_revenue, x='Date', y='Revenue', marker='o', ax=ax[1], color='green')
ax[1].set_title('Tesla Revenue Over Time')
ax[1].set_xlabel('Date')
ax[1].set_ylabel('Revenue (USD in Millions)')
ax[1].grid(True)

plt.tight_layout()
plt.show()
```
